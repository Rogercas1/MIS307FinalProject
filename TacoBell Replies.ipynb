Importing everything good and setting up authorization.

```python
import tweepy
import pandas as pd
from keys import *
import matplotlib.pylab as plt
from textblob import TextBlob

api = tweepy.API(tweepy.OAuth2BearerHandler(bearer_token))
name = '@tacobell'
```

------

First Taco Bell Tweet (1518681018466721792)

```python
tweet_id = '1519755578523979776'

replies=[]
for tweet in tweepy.Cursor(api.search_tweets,q='to:'+name, result_type='recent').items(250):
    if hasattr(tweet, 'in_reply_to_status_id_str'):
        if (tweet.in_reply_to_status_id_str==tweet_id):
            replies.append(tweet)
```

```python
id_list = []
text_list = []
for item in replies:
    text_list.append(item.text)
    id_list.append(item.id)
```

```python
d = {'id':id_list, 'text':text_list}
df = pd.DataFrame(d)

df['Polarity'] = df['text'].apply(lambda x: float(TextBlob(x).sentiment.polarity))
df.loc[:,'Classification'] = 'neutral'
df.loc[df['Polarity'] > 0.03, 'Classification'] = 'positive'
df.loc[df['Polarity'] < -0.03, 'Classification'] = 'negative'
```

```python
fig, ax = plt.subplots()
fig.suptitle(f"Polarity of replies over Taco Bell's Tweet: '1519755578523979776' out of {len(replies)} replies", multialignment='center', wrap=True)
ax.hist(df['Polarity'])
ax.grid(which='major', color='grey', linestyle='--')
ax.set_xlabel('Polarity')
ax.set_ylabel('Amount')
ax;
```
